```python
e_matrix_A = tf.random.uniform([2,2],minval=3,maxval=10,dtype=tf.float32,name="matrixA")
print("Matrix A:{}".format(e_matrix_A))
```

```python
eigen_value,eigen_vectors=tf.linalg.eigh(e_matrix_A)
print("eigen vectors:\n\n{}\n\n \neigen values:{}".format(eigen_vectors,eigen_value))
```
